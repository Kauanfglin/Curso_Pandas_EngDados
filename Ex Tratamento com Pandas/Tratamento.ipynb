```python
import pandas as pd
import numpy as np

df = pd.read_excel('dados_treino_pandas.xlsx')

#df.describe()
#df.info()
df
```

```python
df["quantidades"] = df["quantidades"].fillna(0)
df["quantidades"] = df["quantidades"].map(int)
df["valor_produto"] = df["valor_produto"].fillna(39)
df["produtos"] = df["produtos"].fillna("Computador")
df_analise = df[["produtos", "nome_lojas", "valor_produto","quantidades"]].copy()
df_analise["produtos"] = df_analise["produtos"].str.strip().str.title()
df_analise["nome_lojas"] = df_analise["nome_lojas"].str.strip().str.title()
df_analise["Vendas Totais por Loja"] = df_analise["valor_produto"] * df_analise["quantidades"]
df_analise["nome_lojas"] = df_analise["nome_lojas"].fillna("Loja Central")
df_analise["Vendas Totais por Loja"] = (
    df_analise["Vendas Totais por Loja"]
        .astype(str)
        .str.replace("R$", "", regex=False)
        .str.replace(".", "", regex=False)
        .str.replace(",", ".", regex=False)
        .str.strip()
)

df_analise["Vendas Totais por Loja"] = pd.to_numeric(
    df_analise["Vendas Totais por Loja"],
    errors="coerce"
)
#df_analise = df_analise["Vendas Totais por Loja"].map("R${:,.2f}".format)

```

```python
df_analise

def RegiaodaLoja(Loja):
    if Loja == "Loja Central":
        return "Sao Paulo"
    elif Loja == "Tech Store":
        return "Rio de Janerio"
    elif Loja == "Mega Eletro":
        return "Salvador"
    else:
        return "Mato Grosso"

df_analise["Regiao da Loja"] = df_analise["nome_lojas"].map(RegiaodaLoja)
df_analise
```
