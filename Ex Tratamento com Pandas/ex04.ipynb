```python
import pandas as pd
import numpy as np
df = pd.read_excel("rotina_diaria_pessoas.xlsx")
df
```

```python
df_analise = df.copy()
pd.set_option("display.max_rows",None)
df_analise["Gasto_Estimado"] = np.where(df_analise["Gasto_Estimado"].isnull(), "Nao Calculado", df_analise["Gasto_Estimado"])
df_analise["Horário"] = pd.to_datetime(df_analise["Horário"], format="%H:%M")

df_analise_rotina = (
    df_analise[df_analise["Atividade"] == "Almoço"].groupby("Pessoa")["Horário"].mean()
)
df_analise_rotina = df_analise_rotina.dt.strftime("%H:%M")
df_analise_rotina = df_analise_rotina.reset_index()
df_analise_rotina.rename(columns={"Horário": "Horario_Medio_Almoco"}, inplace=True)

df_Analise_horario_medio_janta = (
    df_analise[df_analise["Atividade"] == "Janta"].groupby("Pessoa")["Horário"].mean()
)
df_Analise_horario_medio_janta = df_Analise_horario_medio_janta.dt.strftime("%H:%M")
df_Analise_horario_medio_janta = df_Analise_horario_medio_janta.reset_index()
df_Analise_horario_medio_janta = df_Analise_horario_medio_janta.rename(columns={"Horário": "Horario_Medio_Janta"})

df_analise_rotina = df_analise_rotina.merge(df_Analise_horario_medio_janta, on="Pessoa", how="left")

df_analise["Horário"] = df_analise["Horário"].dt.time

df_analise_rotina
df_analise["Gasto_Estimado_Num"] = pd.to_numeric(df_analise["Gasto_Estimado"], errors='coerce')
gastoTotal = df_analise.groupby("Pessoa")["Gasto_Estimado_Num"].sum()
gastoTotal = gastoTotal.reset_index()
gastoTotal.rename(columns={"Gasto_Estimado_Num": "Gasto_Total"}, inplace=True)

df_analise_rotina = df_analise_rotina.merge(gastoTotal,on = "Pessoa",how="left")
df_analise_rotina

```
